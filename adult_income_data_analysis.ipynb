### 9. What Is The Distribution of Age Column?

```python
print(df['age'].describe())
```

```python
df['Age Group'] = pd.cut(df['age'], bins=range(0, df['age'].max() + 5, 5))
agee = df['Age Group'].value_counts().sort_index()

ax=agee.plot.bar(figsize=(8, 6))
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Age Distribution")
for i, count in enumerate(agee):
    ax.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=9)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 11. What is The Distribution of  workclass Column?

```python
df['workclass']=df['workclass'].replace('?','Unknown')
```

```python
counts = df['workclass'].value_counts()
for i,j in counts.items():
    print(f"{i}: {j}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6)) 

counts.plot.bar(ax=axes[0])
axes[0].set_title("Workclass Distribution")
axes[0].set_ylabel("Count")
axes[0].set_xlabel("Workclass")
axes[0].tick_params(axis='x', rotation=45)

sns.boxplot(data=df['workclass'], ax=axes[1])
axes[1].set_title("Workclass Outliers")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
