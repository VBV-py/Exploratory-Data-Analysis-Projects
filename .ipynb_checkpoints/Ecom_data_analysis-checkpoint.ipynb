### What Are The Top 5 Most Popular Email providers (e.g. gmail.com, yahoo.com, etc...) 

```python
domain = df['Email'].str.split('@',expand=True)
```

```python
df['Email Providers'] = domain[1]
```

```python
providers = df['Email Providers'].value_counts()
for domain,count in providers.items():
    print(f"{domain}: {count}")
```

```python
top5= providers.head(5)
print(f"Top 5 most popular email providers: ")
for domain,count in top5.items():
    print(f"{domain}: {count}")
```
