9.	What is the average loudness (dB) of songs by decade?

```python
df['decade'] = (df['year'] // 10) * 10
avg=df.groupby('decade')['dB'].mean().round(2)
print(avg)
```

14.	Make a heatmap of correlation between numeric features (bpm, energy, danceability, etc.).

```python
df.select_dtypes(include='number')
```

```python
numeric_cols = ['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'pop']
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')

plt.title('Correlation of Features')
plt.tight_layout()
plt.show()
```

17.	Find the top 3 artists with the most songs in the dataset and calculate their average valence and energy.

```python
top_artist=df['artist'].value_counts().head(3)
print(top_artist)
top_a=df[df['artist'].isin(top_artist.index)]
top_avg=top_a.groupby('artist')[['val','nrgy']].mean()
print(top_avg)
```

18.	Are live songs (liveness > 0.5) more or less popular than non-live songs?

19.	Create a new column called mood_score = val * dnce. Which songs have the highest mood score?

```python
df['mood_score']=df['val']*df['dnce']
high_mood=df[['title','mood_score']].sort_values(by='mood_score',ascending=False)
print(high_mood.head()) #remove index by .to_string(index=False)
```

20.	Are songs with higher speechiness (spch > 0.5) generally less popular?

```python
less_speech_pop = df[df['spch']<=0.5]['pop'].mean()
high_speech_pop = df[df['spch']>0.5]['pop'].mean()

print(high_speech_pop,less_speech_pop)

```

```python
plt.bar(['High Speechiness (>0.5)', 'Low Speechiness (‚â§0.5)'],
        [high_speech_pop, less_speech_pop],
        color=['red', 'green']);
```

21.	Classify each song as:
‚Ä¢	'High Energy' if nrgy > 70
‚Ä¢	'Low Energy' if nrgy < 40
‚Ä¢	'Medium Energy' otherwise
Then count how many songs fall in each category.

```python
def energy_cat(nrgy):
    if nrgy>70:
        return 'High Energy'
    elif nrgy<40:
        return 'Low Energy'
    else:
        return 'Medium Energy'

df['Energy Category']=df['nrgy'].apply(energy_cat)
count_energy=df['Energy Category'].value_counts()
print(count_energy)
```
