```python
#1 IMPORT LIBRARY AND LOAD DATASET
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris(as_frame=True)
df=iris.frame

print(df)
```

```python
df['species']=df['target'].map(dict(zip(range(3),iris.target_names)))
```

```python
#3 Variate ANALYSIS

num_cols= df.select_dtypes(include='number').columns[:-1]
print(num_cols)

for col in num_cols:
    sns.histplot(df[col],kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
sns.countplot(x='species',data=df)
plt.title("Distribution of Species")
plt.show()
```

```python
#5 MULTIVARIATE ANALYSIS

sns.pairplot(df, hue='species')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
for col in num_cols:
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f"{col} by Species")
    plt.show()
```

```python
#6 FEATURE ENG...

df['petal_area'] = df['petal length (cm)'] * df['petal width (cm)']
df['sepal_area'] = df['sepal length (cm)'] * df['sepal width (cm)']

print(df[['petal_area', 'sepal_area']].describe())
```

```python
#7 CORRELATION ANALYSIS

# Correlation matrix
corr = df.select_dtypes(include='number').drop(columns='target', errors='ignore').corr()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
#8 OUTLIERS DETECTION

for col in num_cols:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# df_outlier_removed = df.copy()
print(f"Original shape: {df.shape}")

# Features to check
num_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

# Loop through each numerical column
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Show shape before and after
print(f"After outlier removal: {df.shape}")
print(df.describe())
```

```python
for col in num_cols:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col} after Outlier Removal")
    plt.show()
```

```python

print(df.groupby('species')[num_cols].mean())

# Violin plots for deeper distribution understanding
for col in num_cols:
    sns.violinplot(x='species', y=col, data=df)
    plt.title(f"{col} by Species")
    plt.show()
```

```python
#9 VISUALIZATION

sns.kdeplot(data=df, x='petal length (cm)', hue='species', fill=True)
plt.title("KDE of Petal Length by Species")
plt.show()
```

```python
#10 CLEAned DATA

print("Final shape:", df.shape)
df.to_csv("iris_cleaned_eda.csv", index=False)
```
